```python
import random

name = input(" What is your name? ")
print("Good Luck !", name)

words = ['rainbow', 'cat', 'computer', 'dog', 'science', 'event', 'programming', 'python', 'java', 'mathematics', 'history', 
         'geography', 'player', 'condition', 'perfect', 'forward', 'reverse', 'plan', 'water', 'board', 'greek']

#random.choice() function will choose one random word from the above list.
word = random.choice(words)

print("Guess the characters")

guesses  = ''

#user will have 10 attempts to guess the word.
attempts  = 10

while  attempts > 0:
    
    #counts the number of times the user fails.
    fail = 0
    
    #all the characters from input word taking one at a time.
    for char in word :
        
        #Comaparing that character with the character in guesses
        if char in  guesses :
            print(char)
            
        else :
            print("---")
            
            #For every  wrong Guess, 1 will be incremented in fail
            fail += 1
            
    if fail == 0:
        print("Congratulations, ",name ,"! You won this game.")
        
        print("The word is : ", word)
        break
        
    #If user inputs wrong character, it will ask to  enter another alphabet.
    guess = input("Guess character : ")
    
    #Every input character will be stored in guesses.
    guesses += guess
    
    #Check the entered character in the word
    if guess not in word :
        attempts -= 1
        
        #if character doesn't match, wrong will be given as output
        print("Wrong, Please enter the new character.")
        
        #This will print the number of attempts left
        print("you have ", +attempts, "more attempts left." )
        
        if attempts ==0:
            print("You loose this game.")
```
